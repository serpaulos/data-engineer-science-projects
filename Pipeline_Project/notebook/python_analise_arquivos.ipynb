## Aquisição dos Dados

```python
ACCESS_KEY='tensorflow'
SECRET_KEY='tensorflow'
```

```python
client = Minio(
       '172.25.0.4:9000',
        access_key=ACCESS_KEY,
        secret_key=SECRET_KEY,
        secure=False
    )
```

```python
bucket_name = 'processing'
parque_folder_path = 'parquet_read'
```

```python
def extract():  
    #busca a lista de objetos no data lake.
    objects = client.list_objects('processing', recursive=True)
    
    #faz o download de cada arquivo e concatena com o dataframe vazio.
    for obj in objects:
        print("Downloading files...")
        print(obj.bucket_name, obj.object_name)

        client.fget_object(
                    obj.bucket_name,
                    obj.object_name,
                    'parquet_read/'+obj.object_name,
        )
```

```python
extract()
```

## Criacao dos datasets

```python
# carregandos arquivos de dados
# Dataset de clientes olist_customers_dataset.csv
df_custumers = pd.read_parquet('parquet_saved/olist_customers_dataset.parquet')
# Dataset de geolocalizacao olist_geolocation_dataset.parquet
df_geolocation = pd.read_parquet('parquet_saved/olist_geolocation_dataset.parquet')
# Dataset de itens olist_order_items_dataset.parquet
df_order_itens = pd.read_parquet('parquet_saved/olist_order_items_dataset.parquet')
# Dataset de pagamentos olist_order_payments_dataset.parquet
df_order_payments = pd.read_parquet('parquet_saved/olist_order_payments_dataset.parquet')
# Dataset de avaliacoes olist_order_reviews_dataset.parquet
df_order_reviews = pd.read_parquet('parquet_saved/olist_order_reviews_dataset.parquet')
# Dataset de ordens olist_orders_dataset.parquet
df_orders = pd.read_parquet('parquet_saved/olist_orders_dataset.parquet')
# Dataset de produtos olist_products_dataset.parquet
df_products = pd.read_parquet('parquet_saved/olist_products_dataset.parquet')
# Dataset de vendedores olist_sellers_dataset.parquet
df_sellers = pd.read_parquet('parquet_saved/olist_sellers_dataset.parquet')
# Dataset de categoria de produto product_category_name_translation.parquet
df_product_category = pd.read_parquet('parquet_saved/product_category_name_translation.parquet')
```

```python
# Criar varios dataframe com o resultado dos tipos de cada variavel
df_custumers_dtypes = pd.DataFrame(df_custumers.dtypes, columns=['Olist_Customers'])
df_geolocation_dtypes = pd.DataFrame(df_geolocation.dtypes, columns=['Olist_Geolocation'])
df_order_itens_dtypes = pd.DataFrame(df_order_itens.dtypes, columns=['Olist_Order_Itens'])
df_order_payments_dtypes = pd.DataFrame(df_order_payments.dtypes, columns=['Olist_Order_Payments'])
df_order_reviews_dtypes = pd.DataFrame(df_order_reviews.dtypes, columns=['Olist_Order_Reviews'])
df_orders_dtypes = pd.DataFrame(df_orders.dtypes, columns=['Olist_Orders'])
df_products_dtypes = pd.DataFrame(df_products.dtypes, columns=['Olist_Products'])
df_sellers_dtypes = pd.DataFrame(df_sellers.dtypes, columns=['Olist_Sellers'])
df_product_category_dtypes = pd.DataFrame(df_product_category.dtypes, columns=['Olist_Product_Category'])

# criar dataframe vazio
df_dtypes2 = pd.DataFrame()

# concatena os dataframe em um unico dataframe
df_dtypes = pd.concat([df_custumers_dtypes, df_geolocation_dtypes, df_order_itens_dtypes,
                      df_order_payments_dtypes, df_order_reviews_dtypes, df_orders_dtypes,
                      df_products_dtypes, df_sellers_dtypes, df_product_category_dtypes], axis=0)
```
